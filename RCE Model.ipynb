### Initial look at the RCE model

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
%matplotlib inline
```

```python
from RCEModel import *
model = RCEModel()
model.spinup()
model.run()
```

```python
dset = xr.open_dataset('model_default.spinup.nc', 'r')
```

Plot temperatures and fluxes during spinup

```python
plt.figure()
plt.plot(dset['time']/86400., dset['T_s'], color = 'black', label = '$T_s$')
plt.plot(dset['time']/86400., dset['T_a'], color = 'blue', label = '$T_a$')
plt.xlabel('Time (d)')
plt.ylabel('Temperature (K)')
```

```python
plt.figure()
plt.plot(dset['time']/86400., dset['sw_net_surf'], color = 'blue', label = 'SW net')
plt.plot(dset['time']/86400., dset['lw_down_surf'], color = 'cyan', label = 'LW down')
plt.plot(dset['time']/86400., dset['lw_up_surf'], color = 'magenta', label = 'LW up')
plt.plot(dset['time']/86400., -dset['sfc_flux'], color = 'red', label = 'Surface flux')
plt.xlabel('Time (d)')
plt.ylabel('Flux (W m$^{-2}$)')
plt.legend()
```

```python
plt.figure()
plt.plot(dset['time']/86400., dset['lw_em_atm'], color = 'blue', label = 'LW emission')
plt.plot(dset['time']/86400., dset['lw_abs_atm'], color = 'cyan', label = 'LW absorption')
plt.plot(dset['time']/86400., dset['sfc_flux'], color = 'red', label = 'Surface flux')
plt.xlabel('Time (d)')
plt.ylabel('Flux (W m$^{-2}$)')
plt.legend()
```
